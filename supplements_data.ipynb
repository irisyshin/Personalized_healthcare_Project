```python
import math, requests, pandas as pd
from config import API_KEY # <--- config.py에서 API_KEY를 불러오도록 수정

#API_KEY 본인 인증키
SERVICE = "C003"
BASE = f"https://openapi.foodsafetykorea.go.kr/api/{API_KEY}/{SERVICE}/json"

# 1회에 1000행까지 받는 걸 권장(속도↑, 호출수↓). 필요시 100으로 낮춰도 됨.
PAGE_SIZE = 1000

def fetch_range(start, end):
    url = f"{BASE}/{start}/{end}"
    r = requests.get(url, timeout=20)
    r.raise_for_status()
    data = r.json()

    # 서비스마다 최상위 키가 달라질 수 있는데, 대부분 {SERVICE} 키 안에 row가 있어요.
    # C003 응답 예: data['C003']['row']
    block = data.get(SERVICE, {})
    rows = block.get('row', [])
    return rows

all_rows = []
start = 1
while True:
    end = start + PAGE_SIZE - 1
    rows = fetch_range(start, end)
    if not rows:
        break
    all_rows.extend(rows)
    start = end + 1

df = pd.DataFrame(all_rows)
df.to_csv("foodsafety_C003.csv", index=False, encoding="utf-8-sig")
print("저장완료: foodsafety_C003.csv, 행수 =", len(df))
```

# PK 결정 : PRDLST_REPORT_NO (품목제조번호)

```python
import pandas as pd
import numpy as np

df = pd.read_csv("foodsafety_C003.csv", dtype={"PRDLST_REPORT_NO":"Int64", "LCNS_NO":"Int64"})

# PK 후보 검증: 신고번호가 유일한지 확인
n_total = len(df)
n_unique = df["PRDLST_REPORT_NO"].nunique()
print(f"[체크] 전체 행수 {n_total}, 신고번호 고유값 {n_unique}")
```

```python
# 유일성/결측 확인
na_cnt = df["PRDLST_REPORT_NO"].isna().sum()
dup_cnt = df["PRDLST_REPORT_NO"].duplicated().sum()
print("[체크] 전체행:", n_total, "고유값:", n_unique, "결측:", na_cnt, "중복개수:", dup_cnt)

# 중복 사례가 있으면 몇 건만 보기
if dup_cnt > 0:
    dups = df[df["PRDLST_REPORT_NO"].duplicated(keep=False)].sort_values("PRDLST_REPORT_NO")
    print("중복 예시 상위 5건:\n", dups[["PRDLST_REPORT_NO","PRDLST_NM","LAST_UPDT_DTM"]].head())
```

# 전처리 - 필요없는 컬럼 drop

```python
import re

# 1) 필요한 컬럼만 선택 (있을 때만 자동으로 가져오도록 안전하게)
must_cols = [
    "PRDLST_REPORT_NO",    # 품목제조번호 PK
    "PRDLST_NM",           # 품목명
    "PRIMARY_FNCLTY",      # 주된 기능성
    "RAWMTRL_NM",          # 원재료
    "IFTKN_ATNT_MATR_CN",  # 섭취 시 주의사항
    "NTK_MTHD",            # 섭취방법
    "BSSH_NM",             # 업소명
]
## 보조 컬럼들
nice_to_have = [
    "PRDT_SHAP_CD_NM",     # 제형
    "PRMS_DT",             # 보고일자
    "LAST_UPDT_DTM",       # 최종 수정일시
]
cols = [c for c in (must_cols + nice_to_have) if c in df.columns]
df1 = df[cols].copy()
```

```python
# 2) 날짜형 변환 (YYYYMMDD, YYYYMMDDhhmmss 혼재 방어)
def parse_yyyymmdd(x):
    s = str(x)
    if s.isdigit() and len(s)==8:
        return pd.to_datetime(s, format="%Y%m%d", errors="coerce")
    return pd.NaT

def parse_yyyymmddhhmmss(x):
    s = str(x)
    if s.isdigit() and len(s)==14:
        return pd.to_datetime(s, format="%Y%m%d%H%M%S", errors="coerce")
    if s.isdigit() and len(s)==8:
        return pd.to_datetime(s, format="%Y%m%d", errors="coerce")
    return pd.NaT

if "PRMS_DT" in df1.columns:
    df1["PRMS_DT"] = df1["PRMS_DT"].apply(parse_yyyymmdd)
if "LAST_UPDT_DTM" in df1.columns:
    df1["LAST_UPDT_DTM"] = df1["LAST_UPDT_DTM"].apply(parse_yyyymmddhhmmss)
```

```python
df1.head(5)
```

```python
# 3) 텍스트 클린업 (앞뒤 공백/중복 공백 제거, 비표준 공백 정규화)
def clean_text(s):
    if pd.isna(s): 
        return s
    # 줄바꿈/탭을 공백으로
    s = str(s).replace("\r", " ").replace("\n", " ").replace("\t", " ")
    # 연속 공백 1칸으로 축약
    s = re.sub(r"\s+", " ", s).strip()
    return s

text_cols = [c for c in ["PRDLST_NM","PRIMARY_FNCLTY","RAWMTRL_NM","IFTKN_ATNT_MATR_CN","NTK_MTHD","BSSH_NM","PRDT_SHAP_CD_NM"] if c in df1.columns]
for c in text_cols:
    df1[c] = df1[c].apply(clean_text)
```

```python
# 4) PK 무결성 재확인(안전 차원)
assert df1["PRDLST_REPORT_NO"].isna().sum() == 0, "PK에 결측이 있습니다."
assert df1["PRDLST_REPORT_NO"].duplicated().sum() == 0, "PK 중복이 있습니다."
```

```python
# 5) 컬럼 순서 정리(가독성)
ordered = [c for c in ["PRDLST_REPORT_NO","PRDLST_NM","BSSH_NM","PRDT_SHAP_CD_NM",
                       "PRIMARY_FNCLTY","RAWMTRL_NM","IFTKN_ATNT_MATR_CN","NTK_MTHD",
                       "PRMS_DT","LAST_UPDT_DTM"] if c in df1.columns]
df_clean = df1[ordered].copy()
```

# 전처리 - NULL값 처리

### PRDT_SHAP_CD_NM

```python
mask = df_clean["PRDT_SHAP_CD_NM"].isna()
need_shape = df_clean.loc[mask, ["PRDLST_REPORT_NO","PRDLST_NM","BSSH_NM","RAWMTRL_NM"]]
print(need_shape.shape)     # (5, 4) 예상
need_shape.head(5)          # 품목명/업소명 보고 제형 판단
```

```python
df_clean["PRDT_SHAP_CD_NM"].value_counts()
```

```python
# 직접 채우기
manual_shape = {
    # PRDLST_REPORT_NO: "제형",
    # 202401010123: "정",
    # 202312311234: "캡슐",
}

# 채우기 + '수정 여부' 플래그 남기기
was_na = df_clean["PRDT_SHAP_CD_NM"].isna()
df_clean.loc[df_clean["PRDLST_REPORT_NO"].isin(manual_shape.keys()), "PRDT_SHAP_CD_NM"] = \
    df_clean["PRDLST_REPORT_NO"].map(manual_shape)

df_clean["SHAPE_IMPUTED"] = False
df_clean.loc[was_na & df_clean["PRDT_SHAP_CD_NM"].notna(), "SHAPE_IMPUTED"] = True
```

```python
# 직접 채우기 : 결측치 행을 확인한 정보를 바탕으로 'PRDLST_REPORT_NO'를 사용하여 인덱스를 지정합니다.

# 2004001507353 (장박시다이어트)는 '환'
df_clean.loc[df_clean["PRDLST_REPORT_NO"] == 2004001507353, "PRDT_SHAP_CD_NM"] = "환"

# 200400150831108 (눈에좋은루테인)는 '캡슐'
df_clean.loc[df_clean["PRDLST_REPORT_NO"] == 200400150831108, "PRDT_SHAP_CD_NM"] = "캡슐"

# 2004001700248 (장건강 데일리 팜)은 '캡슐' 
df_clean.loc[df_clean["PRDLST_REPORT_NO"] == 2014001700248, "PRDT_SHAP_CD_NM"] = "캡슐"

# 2004001710734 (프로바이오틱스 HY7714 SC)는 '분말'
df_clean.loc[df_clean["PRDLST_REPORT_NO"] == 2014001710734, "PRDT_SHAP_CD_NM"] = "분말"

# 2006000203528 (울트라원)은 '정' 
df_clean.loc[df_clean["PRDLST_REPORT_NO"] == 20060020003528, "PRDT_SHAP_CD_NM"] = "정"
```
