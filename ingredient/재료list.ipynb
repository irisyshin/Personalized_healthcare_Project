```python
import pandas as pd
import ast  # 문자열 → 딕셔너리로 변환용

# CSV 불러오기
df = pd.read_csv("recipe.csv")

# ingredient_full 컬럼에서 key(재료명)만 추출
all_keys = []

for val in df['ingredient_full'].dropna():
    try:
        d = ast.literal_eval(val)  # 문자열을 안전하게 딕셔너리로 변환
        all_keys.extend(d.keys())  # key(재료명)만 추가
    except Exception:
        continue  # 혹시 JSON 형식 깨진 행은 건너뛰기

# 중복 제거 및 정렬
unique_ingredients = sorted(set(all_keys))

# 결과 확인
print(f"✅ 총 {len(unique_ingredients)}개의 고유 재료명 추출 완료")
print(unique_ingredients)  # 앞부분만 보기

# CSV 저장 (선택)
pd.Series(unique_ingredients, name="INGREDIENT_NAME").to_csv(
    "ingredient_keys_from_recipe.csv", index=False, encoding="utf-8-sig"
)
```

```python
import pandas as pd

# CSV 불러오기
df = pd.read_csv("ingredient_keys_from_recipe.csv")

# 삭제할 키워드 목록
remove_keywords = [
    "양념", "소금", "고추장", "뼈", "간장", "후추", "2배", "1인", "물", "육수",
    "가진", "가쯔오", "각종", "설탕", "전분", "강력분", "걸름망", "기름", "소스", "국물", "가는", "장국", "맛술", 
    "맛간장", "다시다", "다시", "다대기", "다용도", "다용도용", "다용도용도", "양념장", "양념간장", "양념간장용", "양념간장용도",
    "간", "갈은", "가루", "팬", "즙", "부르스타", "볼", ""
]

# 각 키워드가 포함된 행 필터링 제거
pattern = "|".join(remove_keywords)  # 정규식 OR 조건 생성
df_filtered = df.drop_duplicates(subset="INGREDIENT_NAME").reset_index(drop=True)

df_filtered = df_filtered[~df_filtered["INGREDIENT_NAME"].str.contains(pattern, case=False, na=False)]
# 특정 단어('다진') 정확히 일치하는 행 제거
df_filtered = df_filtered[df_filtered["INGREDIENT_NAME"] != "다진"]
# 결과 확인
print(f"✅ 필터링 완료: {len(df)} → {len(df_filtered)}개 남음")
print(df_filtered.head())

# CSV 저장 (선택)
df_filtered.to_csv("ingredient_keys_filtered.csv", index=False, encoding="utf-8-sig")
```
