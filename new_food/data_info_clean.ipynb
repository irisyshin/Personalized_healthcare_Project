```python
# 데이터 불러오기 
import pandas as pd

try:
    df = pd.read_csv('recipe_data_info.csv', encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv('recipe_data_info.csv', encoding='cp949')

print(df.shape)
print(df.columns)
print(df.head())
```

```python
df_clean = df.copy()
```

```python
df_clean = df_clean.drop(columns=['ROW_NUM', 'NATION_CODE', 'TY_CODE', 'PC_NM'], errors='ignore')
```

```python
df_clean.head()
```

```python
# 'Kcal' 문자 제거 후 숫자 변환
df_clean['CALORIE'] = df_clean['CALORIE'].astype(str).str.replace('Kcal', '', case=False).str.strip()
df_clean['CALORIE'] = pd.to_numeric(df_clean['CALORIE'], errors='coerce')

# 인분 수 추출
df_clean['QNT_NUM'] = df_clean['QNT'].astype(str).str.extract(r'(\d+)').astype(float)
df_clean['QNT_NUM'] = df_clean['QNT_NUM'].fillna(1)

# 1인분 기준 칼로리 계산
df_clean['1QNT_CALORIE'] = df_clean['CALORIE'] / df_clean['QNT_NUM']
df_clean = df_clean.drop(columns=['QNT_NUM'], errors='ignore')


print(df_clean[['QNT', 'CALORIE', '1QNT_CALORIE']].head())

```

```python
df_clean
```

```python
# LEVEL_NM 값 치환
df_clean['LEVEL_NM'] = df_clean['LEVEL_NM'].replace({
    '초보환영': '하',
    '보통': '상',
    '어려움': '상'
})

print(df_clean['LEVEL_NM'].unique())
```

```python
df_clean = df_clean.drop(columns=['CALORIE', 'QNT'], errors='ignore')
df_clean
```

```python
print(df_clean['LEVEL_NM'].unique())
```

```python
# '분' 제거 후 숫자형으로 변환
df_clean['COOKING_TIME'] = df_clean['COOKING_TIME'].astype(str).str.replace('분', '', regex=False).astype(float)

print(df_clean[['RECIPE_NM_KO', 'COOKING_TIME']].head())
```

```python
zero_count = (df_clean['CALORIE'] == 0).sum()
print("CALORIE 값이 0인 행 개수:", zero_count)
```

```python
df_clean.to_csv('recipe_info.csv', index=False, encoding='utf-8-sig')
```

```python
from sqlalchemy import create_engine
import pandas as pd

# DB 연결 (비밀번호의 @ → %40 인코딩 주의)
engine = create_engine('mysql+pymysql://lgup3:lgup3P%40ssw0rd@211.51.163.232:19306/lgup3')

# CSV 파일 목록
files = {
    'recipe_info': 'recipe_info.csv',
    'recipe_ingredient_info': 'recipe_ingredient_info.csv',
    'recipe_process_info': 'recipe_process_info.csv'
}

# 각 파일 DB에 업로드
for table, path in files.items():
    df = pd.read_csv(path, encoding='utf-8-sig')
    df.to_sql(table, con=engine, if_exists='replace', index=False)
    print(f"{table} 테이블 업로드 완료 ({len(df)}행)")
```
