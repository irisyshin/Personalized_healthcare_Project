```python
import pandas as pd

# CSV 파일 불러오기
df1 = pd.read_csv('식품영양성분.csv', encoding='utf-8-sig')
df2 = pd.read_csv('추천식품db1정제.csv', encoding='utf-8-sig')

# 같은 컬럼 기준으로 합치기
merged_df = pd.concat([df1, df2], ignore_index=True)

# 결과 저장
merged_df.to_csv('검색식품데이터.csv', index=False, encoding='utf-8-sig')
```

```python
df3 = pd.read_csv('검색식품데이터.csv', encoding='utf-8-sig')
```

```python
pd.set_option('display.max_rows', None)
print(df3['food_name'].value_counts()[lambda x: x > 1])
```

```python
dup_count = (df3['food_name'].value_counts() > 1).sum()
print("중복된 food_name 종류 개수:", dup_count)
```

```python
import pandas as pd

# 1️⃣ df3에서 중복된 food_name만 추출
dup_list = df3['food_name'].value_counts()[lambda x: x > 1].index.tolist()

# 2️⃣ 추천식품 DB 불러오기
rec_df = pd.read_csv('추천식품db1정제.csv', encoding='utf-8-sig')

# 3️⃣ 삭제 전 확인
print("삭제 대상 개수:", len(dup_list))
print("삭제 전 행 수:", len(rec_df))

# 4️⃣ 중복 리스트에 포함된 행 제거
filtered_df = rec_df[~rec_df['food_name'].isin(dup_list)]

# 5️⃣ 결과 저장
filtered_df.to_csv('추천식품db1_중복삭제.csv', index=False, encoding='utf-8-sig')

print("✅ 삭제 완료!")
print("삭제 후 남은 행 수:", len(filtered_df))
```

```python
# CSV 파일 불러오기
df1 = pd.read_csv('식품영양성분.csv', encoding='utf-8-sig')
df2 = pd.read_csv('추천식품db1_중복삭제.csv', encoding='utf-8-sig')

# 같은 컬럼 기준으로 합치기
merged_df = pd.concat([df1, df2], ignore_index=True)

# 결과 저장
merged_df.to_csv('검색식품데이터.csv', index=False, encoding='utf-8-sig')
```
